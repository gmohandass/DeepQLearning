```python
def normalize(obs, time):
    position, velocity = obs
    min_position = -1.2           
    max_position = 0.6
    min_velocity = -0.07           
    max_velocity = 0.07
    min_time = 0
    max_time = 200
    position_norm =  (position-min_position)/(max_position-min_position)
    velocity_norm =  (velocity-min_velocity)/(max_velocity-min_velocity)
    time_norm =  (time-min_time)/(max_time-min_time)
    observations = [position_norm, velocity_norm, time_norm]
    return np.array([observations])

def to_tensor(observations):
    return tf.reshape(tf.convert_to_tensor(observations[0], dtype=tf.float32), (1, observations[0].shape[0]))
```

```python
class ExperiencedReplay(object):
    def __init__(self, buffer_size, batch_size):
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.state_buffer = None
        self.action_buffer = None
        self.reward_buffer = None
        self.new_state_buffer = None
        self.done_buffer = None
        
    def add_exp(self, state, action, reward, new_state, done):
        sars_bundle = [state, action, reward, new_state]
        buffer_size = self.buffer_size
        self.state_buffer = self._concat(self.state_buffer, state)
        self.action_buffer = self._concat(self.action_buffer, np.array([[action]]))
        self.reward_buffer = self._concat(self.reward_buffer, np.array([[reward]]))
        self.done_buffer = self._concat(self.done_buffer, np.array([[done]]))
        self.new_state_buffer = self._concat(self.new_state_buffer, new_state)
        
        if self.state_buffer.shape[0] > buffer_size:
            self.state_buffer = self.state_buffer[:buffer_size, :]
            self.action_buffer = self.action_buffer[:buffer_size, :]
            self.reward_buffer = self.reward_buffer[:buffer_size, :]
            self.done_buffer = self.done_buffer[:buffer_size, :]
            self.new_state_buffer = self.new_state_buffer[:buffer_size, :]
            
    def _concat(self, old_array, new_array):
        return np.concatenate([new_array, old_array], axis=0) if old_array is not None else new_array
    
    def _tf_tensor(self, np_array):
        return tf.random.shuffle(tf.convert_to_tensor(np_array, dtype=tf.float32))
        
    def sample(self):
        if self.state_buffer.shape[0] > self.batch_size:
            idx = np.random.choice(self.state_buffer.shape[0], self.batch_size, replace=False)
        else:    
            idx = list(range(self.state_buffer.shape[0]))
        
        sample_state = self._tf_tensor(self.state_buffer[idx, :])
        sample_action = self._tf_tensor(self.action_buffer[idx, :])
        sample_reward = self._tf_tensor(self.reward_buffer[idx, :])
        sample_done = self._tf_tensor(self.done_buffer[idx, :])
        sample_new_state = self._tf_tensor(self.new_state_buffer[idx, :])
         
        return sample_state, sample_action, sample_reward, sample_new_state, sample_done
        
```

```python
import numpy as np
import gym
import random

env = gym.make('MountainCar-v0')
total_episodes = 100
C = 10
num_inputs, num_actions = env.observation_space.shape[0] + 1, env.action_space.n
learning_rate = 0.001
q_network = define_q_network(num_inputs, num_actions, learning_rate)
target_q_network = define_q_network(num_inputs, num_actions, learning_rate)
gamma = 0.99
replay_buffer = ExperiencedReplay(20000, 128)
```

```python
epsilon = 1.0
epsilon_decay = 0.001
for i_episode in range(total_episodes):
    observation = env.reset()
    current_state = normalize(observation, 0)
    rewards = []
    for t in range(1, 200):
                
        q_pred = q_network(to_tensor(current_state))
        epsilon = get_epsilon(i_episode, epsilon, epsilon_decay)
        if random.random() < epsilon:
            action_type = "Explore"
            action = env.action_space.sample()
        else:
            action_type = "Exploit"
            action = tf.argmax(q_pred, axis=1).numpy()[0]

        observation, reward, done, info = env.step(action)
        reward = reward * t  + observation[1] + 10 * abs(observation[0])
        new_state = normalize(observation, t)

        replay_buffer.add_exp(current_state, action, reward, new_state, done)

        current_state = new_state
        rewards.append(reward)
            
        sample_state, sample_action, sample_reward, sample_new_state, sample_done = replay_buffer.sample()
        sample_actions_encoded = tf.squeeze(tf.one_hot(tf.cast(sample_action, tf.int32), num_actions, on_value=1.0, off_value=0.0))
        
        future_q_val = tf.expand_dims(target_q_network.predict(sample_new_state).max(axis=1), axis=1)        
        target_q = sample_reward + (tf.constant(gamma, dtype=tf.float32) * tf.cast(future_q_val, tf.float32) * (1 - sample_done))
        
        with tf.GradientTape() as tape:    
            q_pred = tf.reduce_sum(q_network(sample_state) * sample_actions_encoded, axis=-1)
            loss = tf.keras.losses.MSE(target_q, q_pred)
            # loss = tf.keras.losses.Huber()(target_q, q_pred)
        
        grads = tape.gradient(loss, q_network.trainable_weights)
        q_network.optimizer.apply_gradients(zip(grads, q_network.trainable_weights))
        
        if (i_episode)%C == 0 & t == 1:
            print("Updated target Q network")
            target_q_network.set_weights(q_network.get_weights())
        
        
        show_state(env, action, action_type, epsilon, rewards, episode=i_episode, step=t)
        if done:
            if reward == 0:
                print(f"Success {i_episode}")
            print(f"Episode {i_episode} finished after {t} time steps with total reward of {sum(rewards)}")
            break
        
env.close()
```
